# store data into array

```python
import sys
import numpy as np

filename = "cities.csv" #it can be .txt or .csv 
X = []
y = []

with open(filename, 'r') as f:
    for line in f.readlines():
        data = [(i) for i in line.split(',')]
        print(data)
        #split and store data in arrays as you wish
        xt = data[:-1] #just discard last
        yt = data[-1] #use only last
        X.append(xt)
        y.append(yt)
```
